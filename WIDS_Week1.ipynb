```python
# Reshape the training and test datasets to include a single channel (for grayscale images)
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)  # Reshape to (num_samples, height, width, channels)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
input_shape = (28, 28, 1)  # Define the input shape for the model

# Convert data type to float32 to allow decimal values after normalization
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# Normalize pixel values to the range [0, 1] by dividing by the maximum RGB value (255)
X_train /= 255
X_test /= 255

# Print the shape of the training data and the number of images in the datasets
print('x_train shape:', X_train.shape)  # Shape of the reshaped training data
print('Number of images in x_train:', X_train.shape[0])  # Total training images
print('Number of images in x_test:', X_test.shape[0])  # Total test images
```

```python
# specify the maximum rotation_range angle
rotation_range_val = 30

# import relevant library
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# create the class object
datagen = ImageDataGenerator(rotation_range=rotation_range_val)

# fit the generator
datagen.fit(X_train.reshape(X_train.shape[0], 28, 28, 1))

# define number of rows & columns
num_row = 2
num_col = 8
num= num_row*num_col

# plot before
print('BEFORE:\n')
# plot images
fig1, axes1 = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
for i in range(num):
    ax = axes1[i//num_col, i%num_col]
    ax.imshow(X_train[i], cmap='gray_r')
    ax.set_title('Label: {}'.format(Y_train[i]))
plt.tight_layout()
plt.show()

# plot after
print('AFTER:\n')
fig2, axes2 = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
for X, Y in datagen.flow(X_train.reshape(X_train.shape[0], 28, 28, 1),
                         Y_train.reshape(Y_train.shape[0], 1),
                         batch_size=num,
                         shuffle=False):
    for i in range(0, num):
        ax = axes2[i//num_col, i%num_col]
        ax.imshow(X[i].reshape(28,28), cmap='gray_r')
        ax.set_title('Label: {}'.format(int(Y[i])))
    break
plt.tight_layout()
plt.show()
```

```python
# Specify the augmentation parameters
rotation_range_val = 30  # Range for random rotations
width_shift_val = 0.25  # Fraction of total width for horizontal shifts
height_shift_val = 0.25  # Fraction of total height for vertical shifts
shear_range_val = 45  # Shear angle in degrees
zoom_range_val = [0.5, 1.5]  # Range for random zoom

# Import the required library
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create an ImageDataGenerator object with the specified augmentations
datagen = ImageDataGenerator(rotation_range=rotation_range_val,
                             width_shift_range=width_shift_val,
                             height_shift_range=height_shift_val,
                             zoom_range=zoom_range_val)

# Fit the data generator to the training dataset
datagen.fit(X_train.reshape(X_train.shape[0], 28, 28, 1))

# Define the number of rows and columns for the grid display
num_row = 4  # Number of rows of images
num_col = 8  # Number of columns of images
num = num_row * num_col  # Total number of images to display

# Display original images (BEFORE augmentation)
print('BEFORE:\n')
fig1, axes1 = plt.subplots(num_row, num_col, figsize=(1.5 * num_col, 2 * num_row))
for i in range(num):
    ax = axes1[i // num_col, i % num_col]  # Access the appropriate subplot
    ax.imshow(X_train[i], cmap='gray_r')  # Display the image in grayscale
    ax.set_title('Label: {}'.format(Y_train[i]))  # Show the label of the image
    ax.axis('off')  # Remove axis ticks
plt.tight_layout()  # Adjust spacing between subplots
plt.show()

# Display augmented images (AFTER augmentation)
print('AFTER:\n')
fig2, axes2 = plt.subplots(num_row, num_col, figsize=(1.5 * num_col, 2 * num_row))
for X, Y in datagen.flow(X_train.reshape(X_train.shape[0], 28, 28, 1), 
                         Y_train.reshape(Y_train.shape[0], 1), 
                         batch_size=num, 
                         shuffle=False):  # Generate augmented images in batches
    for i in range(num):
        ax = axes2[i // num_col, i % num_col]  # Access the appropriate subplot
        ax.imshow(X[i].reshape(28, 28), cmap='gray_r')  # Display the augmented image
        ax.set_title('Label: {}'.format(int(Y[i])))  # Show the label of the image
        ax.axis('off')  # Remove axis ticks
    break  # Stop after displaying the first batch
plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
